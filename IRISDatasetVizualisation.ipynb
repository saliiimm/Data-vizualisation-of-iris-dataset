# Vizualization with matplotlib and seaborn

```python
#now lets vizualise using a histogramm the sepal lenght accross our 3 species
species = np.array(iris_dataset['species'].unique())
sepal_l = np.array(iris_dataset['sepal_length'])

hist = plt.hist(sepal_l,bins = 10)
```

```python
#it is still very flou,lets use seaborn to make it better
sea.histplot(iris_dataset,x = iris_dataset['sepal_length'],bins = 10,hue = 'species' )
```

```python
#we can understand that most of the setosa are small,virginaca are long,and versicolor are in the middle
#now lets use boxplots to see the distribution of petal widths for each species
petal_w_setosa = np.array(iris_dataset['petal_width'][iris_dataset['species'] == 'setosa'])
petal_w_setosa
```

```python
peatl_l_versicolor = np.array(iris_dataset['petal_width'][iris_dataset['species'] == 'versicolor'])
peatl_l_versicolor
```

```python
petal_l_virginica = np.array(iris_dataset['petal_length'][iris_dataset.species == 'virginica'])
petal_l_virginica
```

```python
data = [petal_w_setosa,peatl_l_versicolor,petal_l_virginica]
fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])

ax.boxplot(data)
```

```python
#lets see better with seaborn
sea.boxplot(iris_dataset,x = 'species',y = 'petal_length',hue = 'species')
```

```python
#we just confirmed our conclusion for histograms but for lenghts
#lets see the relation between sepal lenght and sepal width using seaborn
sea.scatterplot(iris_dataset,y = 'sepal_width',x = 'sepal_length',hue = 'species')
```

```python
#lets do the same this time for petal
sea.scatterplot(iris_dataset,x = 'petal_length',y = 'petal_width',hue = 'species')
```

```python
#here it is very clear that each spece has a defined lenght and width:
#setosa are small,while versicolor are a bit longer,virginnica is the largest type of iris
sea.pairplot(iris_dataset,hue = 'species')
```

```python
#lets test swarmplot
sea.swarmplot(iris_dataset,x = 'sepal_length',y='species')
```

```python
sea.swarmplot(iris_dataset,x = 'sepal_width',y='species')
```

```python
#now violin plot
sea.violinplot(iris_dataset,x = 'petal_width',y='species')
```

```python
#now jointplot
sea.jointplot(iris_dataset,x = 'petal_width',y = 'petal_length',hue = 'species')
```
